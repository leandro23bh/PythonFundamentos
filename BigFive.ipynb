```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from io import open
pd.options.display.max_columns = 150
```

```python
#Carregando os dados

data = pd.read_csv('/content/data-final.csv', sep='\t')
```

```python
data.describe()
```

```python
# Pegar apenas os dados importante, que so as colunas das perguntas
data.drop(data.columns[50:110], axis=1, inplace=True)
```

```python
pd.options.display.float_format = "{:.2f}".format
data.describe()
```

```python
data["EXT1"].value_counts()
```

```python
#Verifica onde temos os valores zerados.
data["EXT1"].value_counts()
```

```python
#Remove os valores zerados
data = data[(data > 0.00).all(axis=1)]
```

```python
#Elbow Method - Verificar a quantidade de clusters do agrupamento
#Biblioteca de extenso do skLearn
!pip install yellowbrick
```

```python
#Importa o sklearn e o metodo Elbow para verificao de clusters
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
```

```python
kmeans = KMeans()
visualizer = KElbowVisualizer(kmeans, k=(2,10))
```

```python
#Pegar apenas 5000 registros
data_sample = data.sample(n=5000, random_state=1)
```

```python
#Treinar o modelo com o dados de exemplo, para vermos a quantidade ideal de clusters
visualizer.fit(data_sample)
visualizer.poof(legend='true')
```

```python
kmeans = KMeans(n_clusters=5)
k_fit = kmeans.fit(data)
```

```python
#Adicionar os labels no dataframe
predicoes = k_fit.labels_
data['Clusters'] = predicoes
```

```python
#Contagem por grupo(clusters) identificado
data["Clusters"].value_counts()
```

```python
col_list = list(data)
ext = col_list[0:10]
est = col_list[10:20]
agr = col_list[20:30]
csn = col_list[30:40]
opn = col_list[40:50]
```

```python
data_soma = pd.DataFrame()
data_soma['extroversion'] = data[ext].sum(axis=1)/10
data_soma['neurotic'] = data[est].sum(axis=1)/10
data_soma['agreeable'] = data[agr].sum(axis=1)/10
data_soma['conscientious'] = data[csn].sum(axis=1)/10
data_soma['open'] = data[opn].sum(axis=1)/10
data_soma['Clusters'] = predicoes
```

```python

data_soma.groupby('Clusters').mean()
```

```python
data_clusters = data_soma.groupby('Clusters').mean()
```

```python
plt.figure(figsize=(22,3))
for i in range(0, 5):
    plt.subplot(1,5,i+1)
    plt.bar(data_clusters.columns, data_clusters.iloc[:, i], color='green', alpha=0.2)
    plt.plot(data_clusters.columns, data_clusters.iloc[:, i], color='red')
    plt.title('Grupo ' + str(i))
    plt.xticks(rotation=45)
    plt.ylim(0,4);
```
