```python
import librosa
#import librosa.display
import matplotlib.pyplot as plt
```

```python
#Utilizar o audio da propria biblioteca para estudo
Audio = librosa.example('nutcracker')
#Carrega o audio
Matriz,Som = librosa.load(Audio)
```

```python
#Conversaão do audio para Matriz
#Cada valor representa uma onda
Matriz
```

```python
#Verificando a dimensãodos dados das ondas
Matriz.shape
```

```python
#taxa de amostragem do som
Som
```

```python
#Mudando o comportamento do audio para 44,1 khz
librosa.load(Audio, sr=44100)
```

```python
#Plotar a matriz e entender a escala do som
plt.figure(figsize=(14,5))
librosa.display.waveplot(Matriz, sr=Som);
```

```python
#Espectograma
Sinal = librosa.stft(Matriz)

Decibeis = librosa.amplitude_to_db(abs(Sinal))
plt.figure(figsize=(14,5))

librosa.display.specshow(
    Decibeis,
    sr=Som,
    x_axis = 'time',
    y_axis = 'hz'
)
plt.colorbar()
```
