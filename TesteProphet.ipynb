```python
import pandas as pd
from prophet import Prophet
```

```python
df_ = pd.read_csv("/content/testeProphet.csv",sep=';')
```

```python
df_.head()
```

```python
treino = Prophet()
```

```python
treino.fit(df_)
```

```python
future = treino.make_future_dataframe(periods=1460)
future.tail()
```

```python
forecast = treino.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
fig1 = treino.plot(forecast)
```

```python
fig2 = treino.plot_components(forecast)
```

```python
from prophet.plot import plot_plotly, plot_components_plotly

plot_plotly(treino, forecast)
```

```python
plot_components_plotly(treino, forecast)
```
